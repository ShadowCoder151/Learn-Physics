```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## Units
#### Distance: (m), Time(s)

```python
u = 0 # Initial velocity
a = 2 # Acceleration
t = np.arange(0, 10.1, 0.1)  # Time range
```

```python
# Final velocity for each time step
v = u + a * t
```

```python
# Distance for each time step
x = u * t + 0.5 * a * (t ** 2)
```

```python
data = pd.DataFrame({"Time(s)":t, "Displacement(m)":x, "Velocity(m/s)":v, "Acceleration(m/s**2)": a})
```

## Plotting the Graph

### 1. Velocity against time

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x="Time(s)", y="Velocity(m/s)")
plt.title("Velocity vs Time")
plt.xlabel("t(s)")
plt.ylabel("v (m/s)")
plt.grid(True)
plt.show()
```

### 2. Acceleration against time

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x="Time(s)", y="Acceleration(m/s**2)")
plt.title("Acceleration vs Time")
plt.xlabel("t(s)")
plt.ylabel("a (m/s**2)")
plt.grid(True)
plt.show()
```

### 3. Displacement against time

```python
plt.figure(figsize=(10, 6))
sns.lineplot(data=data, x="Time(s)", y="Displacement(m)")
plt.title("Displacement vs Time")
plt.xlabel("t(s)")
plt.ylabel("x(m)")
plt.grid(True)
plt.show()
```
